### In this part, we will transfer a MIDI music (no jazz style) to jazz music by using the well trained NMT model.

```python
import music21
from music21 import  *
import shutil
import pandas as pd
import numpy as np
import sys, re, itertools, random, os
from collections import Counter, defaultdict
from sklearn.cluster import KMeans
import os
from functions import *
```

## Step 1. MIDI --> TXT
In this part, we will input a MIDI music, and use ***TransferMidToChordTxt*** function to transfer the MIDI music to TXT with it track information.

```python
MIDIname = "Despacito"
file_path = "transfer-music/" + MIDIname + ".mid"
TransferMidToChordTxt(file_path)
```

## Step 2. TXT --> Chord & Note
In this step, we split the txt file in different parts by using ***ExtractNoteAndChord***ã€€function, then get the chord and note information.

```python
file_path = "transfer-music/" + MIDIname + "_chord.txt"
split_length = 20
out_note_path = "transfer-music/" + MIDIname + "_chord.note"
out_chord_path = "transfer-music/" + MIDIname + "_chord.chord"
ExtractNoteAndChord(file_path, split_length,out_note_path,out_chord_path)
```

We need to remove the empty lines in files to make sure these files suit the NMT's requirment.

```python
file_path = "transfer-music/" + MIDIname + "_chord.note"
RemoveEmptyLine(file_path)
```

```python
file_path = "transfer-music/" + MIDIname + "_chord.chord"
RemoveEmptyLine(file_path)
```

Now we copy the generate chord file to nmp/tmp folder.

```python
shutil.copy("transfer-music/" + MIDIname + "_chord.chord","nmt/tmp/" + MIDIname + "_chord.chord")
```

## Step 3. Chord --> NMT --> Jazz.Note

Now we use this command in terminal to use the well-trained nmt to transfer the music style

```python
print ("Generate the command:")
print ("python -m nmt.nmt     --out_dir=nmt/tmp/model-data-2    --inference_input_file=nmt/tmp/" + MIDIname +"_chord.chord     --inference_output_file=nmt/tmp/" + MIDIname + "_jazz.note")
```

## Step 4.  Jazz.Note & Original MIDI --> Music21 --> Jazz.MIDI
In this step, we copy the genetated jazz.note to our original folder.

```python
shutil.copy("nmt/tmp/" + MIDIname + "_jazz.note","transfer-music/" + MIDIname +"_jazz.note")
```

Then we use the original MIDI music with the jazz.note to generate a jazz-style song.

```python
midiFilePath = "transfer-music/" + MIDIname +".mid"
translatedChordListFile = "transfer-music/" + MIDIname +"_jazz.note"
outputFile = "transfer-music/" + MIDIname +"_jazz.mid"
TransferStyle(midiFilePath,translatedChordListFile,outputFile)
```
